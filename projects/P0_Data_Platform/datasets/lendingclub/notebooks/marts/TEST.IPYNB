```python
from pathlib import Path
import duckdb
import pandas as pd

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)
pd.set_option("display.width", None)

DB_PATH = (
    Path.cwd()
    / "projects"
    / "P0_Data_Platform"
    / "datasets"
    / "lendingclub"
    / "data"
    / "warehouse"
    / "duckdb"
    / "lendingclub.duckdb"
)

print(DB_PATH)

conn = duckdb.connect(str(DB_PATH))
```

```python
print(
    conn.execute(
        """
        SELECT

            SUM(
                CASE
                    WHEN dti < 0
                    THEN 1
                    ELSE 0
                END
            ) AS negative_dti_rows,

            SUM(
                CASE
                    WHEN dti IS NULL
                    THEN 1
                    ELSE 0
                END
            ) AS null_dti_rows

        FROM feature.loan_features_v1
        """
    ).fetchdf()
)
```
